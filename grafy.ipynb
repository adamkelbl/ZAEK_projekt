# Vizualizace dat

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("porsche_data.xlsx","porsche_data")
df.drop(columns=["Model"], inplace=True)
print(df.describe())
```

```python
# Počet aut podle série
sns.countplot(x='Series', data=df, order=[911, 964, 993, 996, 997, 991, 992], hue="Series", legend=False, palette="Set2")
plt.title('Počet aut podle série', fontsize=14)
plt.xlabel('Model', fontsize=12)
plt.ylabel('Počet vozidel', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Cenová distribuce podle série
sns.boxplot(x="Series", y='Price', data=df, order=[911, 964, 993, 996, 997, 991, 992], hue="Series", legend=False, palette="Set2")
plt.title('Cenová distribuce podle série', fontsize=14)
plt.xlabel('Série', fontsize=12)
plt.ylabel('Cena (v EUR)', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Scatterplot - snaha o zjistění vztahu mezi nájezdem a cenou vozu
# Vztah mezi cenou a nájezdem se jeví jako logaritmický
sns.scatterplot(data=df, x="Mileage", y='Price', color='blue')
plt.title('Scatterplot ceny a nájezdu')
plt.xlabel('Nájezd [km]')
plt.ylabel("Cena")
plt.grid(True)
plt.show()
```

```python
# Scatterplot - cena a výkon
sns.scatterplot(data=df, x="Power", y='Price', color="red")
plt.title('Scatterplot ceny a výkonu')
plt.xlabel("Výkon [hp]")
plt.ylabel("Cena")
plt.grid(True)
plt.show()
```
